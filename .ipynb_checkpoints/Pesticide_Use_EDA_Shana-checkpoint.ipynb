```python
import pandas as pd
import geopandas as gpd
import os
import pyarrow
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use('fivethirtyeight')   
path = "/srv/data/my_shared_data_folder/cpr/"
pd.set_option('display.max_columns', None)
```

```python
pur2020_ag_data = gpd.read_parquet("/srv/data/my_shared_data_folder/cpr/pur2020_ag.parquet",
                                   columns = ["county_name","prodno","use_no", "geometry"])
                                              #"use_no","county_name","acre_treated","COUNTY_CD","site_name","NAMELSAD","COUNTYFP"])
#pur2020_ag_data = pd.read_parquet("/srv/data/my_shared_data_folder/cpr/another_copy_pur2020_ag.parquet", engine="fastparquet") 
```

```python
#pur2020_ag_data = pur2020_ag_data.sample(frac = 0.001)
#list(pur2020_ag_data.columns)
pur2020_ag_data.to_file("pur2020_geometry.geojson", driver="GeoJSON")
```

```python
pur2020_ag_data['prodno'].apply(str)
distinct_pesticides = pur2020_ag_data.prodno.unique()
len(distinct_pesticides)
```

```python
pest_type = pur2020_ag_data.groupby("prodno")["lbs_prd_used"].sum()
pest_type = pest_type.reset_index().rename(columns={'lbs_prd_used':'total_amt_used_(lbs)'})
pest_type
```

```python
pest_use_county = pur2020_ag_data.groupby("county_name")["prodno"].count().sort_values()
pest_use_county = pest_use_county.reset_index()
pest_use_county

plt.figure(figsize=(7,13)) 
plt.barh(y = pest_use_county['county_name'], width= pest_use_county['prodno'])
plt.xticks(rotation=270)                           # Rotate x-labels by 270 degrees
plt.xlabel('Number of Pesticides')                      # x-label added
plt.ylabel('County Name')          # y-label added
plt.title('Number of Pesticides Used by County')  # title added
plt.show()
```

```python
pest_amt_county = pur2020_ag_data.groupby("county_name")["lbs_prd_used"].sum().sort_values()
pest_amt_county = pest_amt_county.reset_index()
pest_amt_county

plt.figure(figsize=(7,13)) 
plt.barh(y = pest_amt_county['county_name'], width = pest_amt_county['lbs_prd_used'])
plt.xticks(rotation=270)                           # Rotate x-labels by 270 degrees
plt.xlabel('Amount of Pesticide (lbs)')                      # x-label added
plt.ylabel('County Name')          # y-label added
plt.title('Amount of Pesticide Used by County')  # title added
plt.show()
```

```python
acre_treated_county = pur2020_ag_data.groupby("county_name")["acre_treated"].sum()
acre_treated_county = pur2020_ag_data.groupby("county_name")["acre_treated"].sum()
acre_treated_county = acre_treated_county.reset_index()
acre_treated_county

plt.figure(figsize=(7,11)) 
plt.barh(y = acre_treated_county['county_name'], width= acre_treated_county['acre_treated'])
plt.xticks(rotation=270)                           # Rotate x-labels by 270 degrees
plt.xlabel('Total Acres Treated')                      # x-label added
plt.ylabel('County Name')          # y-label added
plt.title('Total Acres Treated by County')  # title added
plt.show()
```

```python
# Finding Common columns
a = np.intersect1d(cali_counties.columns, cali_county_cleaned.columns)
 
# Printing common columns
print ("Common Columns:",a)
```
