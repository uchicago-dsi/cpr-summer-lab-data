```python
import os
import pandas as pd
import geopandas as gpd
import json
import matplotlib.pyplot as plt
import fiona
import contextily
pd.set_option('display.max_columns', None)

path = "/srv/data/my_shared_data_folder/cpr/"
```

```python
dir_list = os.listdir(os.getcwd())
directory = pd.Series(dir_list)
print(dir_list)
```

```python
def Merger(county):
    if (directory.eq(f'{county}_pd.parquet')).any() and (directory.eq(f'{county}_geo.parquet')).any() == True:
        countyhold = pd.read_parquet(f'./{county}_pd.parquet', engine = 'fastparquet')
        countyhold_geo = gpd.read_parquet(f'./{county}_geo.parquet')
        
        countyhold_geo.drop(labels = ['county_name'],axis = 1,inplace = True)
        
        countyhold.drop(labels = ['geometry'], axis = 1 ,inplace = True)
        
        merged = countyhold_geo.merge(countyhold, on = 'identifier')
        
        merged.to_parquet(f'{county}_complete.parquet')
    else:
        print('nope')
    return None
```

```python
Merger('Mendocino')
```
